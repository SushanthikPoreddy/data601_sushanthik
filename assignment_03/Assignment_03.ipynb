Generating the Random data based on the input given by the user and loading that data into the CSV file.

```python
from faker import Faker
import random
import csv
faker = Faker()
n = int(input("Enter number of Entities to be generated:"))
for i in range(n):
    def email():
        return faker.email()

    def phonenumbers():
        return faker.phone_number()

    def address():
        return faker.address()

    def name():
        return faker.name()

    def yearborn():
        return faker.date_of_birth()

    def Number_of_kids():
        return random.randrange(5)

    def Categorical_Value():
        return random.choice(["Rent","Own"])

    def Annual_Income():
        return random.randrange(100000)

    def Number_of_Speeding_Tickets():
        return random.randrange(15)

def generate_data():
    return [email(), phonenumbers(), address(), name(), yearborn(), Number_of_kids(), Categorical_Value(), Annual_Income(), Number_of_Speeding_Tickets()]


with open('final_data2.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['email', 'phonenumbers', 'address', 'name','yearborn','Number_of_kids','Categorical_Value','Annual_Income','Number_of_Speeding_Tickets'])
    for i in range(n):
        writer.writerow(generate_data())
```

Generating the Random data based on the input given by the user and loading that data into the JSON file.

```python
from faker import Faker
import random
import json
faker = Faker()
a = []
n = int(input("Enter number of Entities to be generated:"))
for i in range(n):
    data_list = {"email": faker.email(),
                "phonenumbers": faker.phone_number(),
                 "address": faker.address(),
                 "name": faker.name(),
                 "yearborn": faker.date_of_birth(),
                 "Number_of_kids":random.randrange(5),
                 "Categorical_Value":random.choice(["Rent","Own"]),
                 "Annual_Income": random.randrange(100000),
                 "Number_of_Speeding_Tickets":random.randrange(15)
    }

    #print(data_list)
    a.append(data_list)
#print(a)
with open('final_data.json', 'w') as json_file:
    for i in a:
        json.dump(i, json_file,indent = 4,default=str)
    
```
